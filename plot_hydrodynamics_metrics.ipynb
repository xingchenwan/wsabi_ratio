```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
color_plate = ['r', 'b','orange','cyan']
i = 0
fig = plt.figure(figsize=[8,4])
for name, path in paths.items():
    color = color_plate[i % len(color_plate)]
    data = pd.read_csv(paths[name])
    xaxis = data['samples']
    plt.plot(xaxis, data['rmse'], color=color, marker='o', label=name)
    i += 1
plt.xscale('log')
plt.ylim([0.8, 10])
plt.yscale('log')
plt.axhline(1.13, linestyle='--', label='MAP')
plt.legend()
plt.xlabel('Number of samples')
plt.ylabel('RMSE')
```

```python
i = 0
fig = plt.figure(figsize=[8,4])
for name, path in paths.items():
    color = color_plate[i % len(color_plate)]
    data = pd.read_csv(paths[name])
    xaxis = data['samples']
    plt.plot(xaxis, -data['ll'], color=color, marker='o', label=name)
    i += 1
plt.xscale('log')
plt.axhline(227.2, linestyle='--', label='MAP')
plt.legend()
plt.xlabel('Number of samples')
plt.ylabel('Negative Log-likelihood')
```
