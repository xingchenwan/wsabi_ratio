```python
paths = {'WsabiNBQ': '/Users/xingchenwan/Dropbox/4YP/Codes/wsabi_ratio/output/wsabinbq_lml.csv'}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
color_plate = ['r', 'b','orange','cyan']
i = 0
for name, path in paths.items():
    color = color_plate[i % len(color_plate)]
    data = pd.read_csv(paths[name])
    xaxis = np.arange(0, data.shape[0])
    plt.plot(xaxis, data.iloc[:, 1], color=color, label=name)
    i += 1
plt.xscale('log')
plt.ylim([-1000, -200])
plt.legend()
plt.xlabel('Number of samples')
plt.ylabel('Log Marginal Likelihood')
```
